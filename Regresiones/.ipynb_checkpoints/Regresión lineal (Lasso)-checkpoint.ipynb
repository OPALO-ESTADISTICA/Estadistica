```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.linear_model import LassoCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, median_absolute_error, mean_squared_error
```

## 1. Importar conjunto de datos

```python
# Lectura de los datos

loc = r"../Datos/winequality-red.csv" # Ubicación del archivo
data = pd.read_csv(loc,sep=";",)      # Importar datos usando Pandas

names = list(data.columns.values) # Variable nombres
y = list(data["alcohol"])         # Etiqueta de salida
X = (data.drop(["alcohol","quality"],1))  # Variables independientes   
```

## 2. Normalización de variables

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, 
                                                 random_state=10)     # Fraccionar conjuntos de prueba y entrenamiento

# Normalización de los atributos

scaler = preprocessing.StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## 3. Cálculo de alpha y ajuste de regresión

```python
# Ajustar modelo de regresión con datos de entrenamiento
wineModel = LassoCV(cv=10).fit(X_train_scaled,y_train) # Cv - Con validación cruzada se evalúan 10 regresiones (en este caso). Ajusta

# Visualizar curvas de error en validación

plt.figure(figsize=(10,7))
plt.plot(wineModel.alphas_, wineModel.mse_path_,":") # Valores de Alpha vs Error
plt.plot(wineModel.alphas_, wineModel.mse_path_.mean(axis=-1), label="MSE promedio",linewidth=2) # Promedio de errores
plt.axvline(wineModel.alpha_, linestyle="--", label="Estimación del mejor Alpha por CV")
plt.semilogx()
plt.legend(fontsize=12)

ax = plt.gca()
ax.invert_xaxis()

plt.xlabel("Alpha", fontsize=15)
plt.ylabel("Error Cuadrático Medio (MSE)")
plt.axis("tight")
plt.show()
```

```python
print("El valor de alpha que minimiza el error CV es: ", wineModel.alpha_)
print("Minímo MSE:", min(wineModel.mse_path_.mean(axis=-1)))
```

## 4. Evaluación

```python
pred = wineModel.predict(X_test_scaled) # Se pronostican valores para conjunto de prueba
```

```python
# Métricas de evaluación

R_2 = r2_score(y_test,pred)                         # Coeficiente de determinación
MAE = median_absolute_error(y_test,pred)            # Error absoluto medio
RMSE = np.sqrt(mean_squared_error(y_test,pred))     # Raíz del Error cuadrático medio

print("R^2: %.4f" % np.mean(R_2))
print("RMSE: %.4f" % np.mean(RMSE))
print("MAE: %.4f" % np.mean(MAE))
```

```python
# Se presenta visualmente la comparación de 50 observaciones de prueba

plt.figure(figsize=(10,6))
plt.plot(y_test[50:100], label="Observado")
plt.plot(pred[50:100], label="Pronosticado")
plt.xlabel("Vino", fontsize=12)
plt.ylabel("Alcohol", fontsize=12)
plt.legend(fontsize=12)
plt.show()
```
