# Visualización de variables categóricas y cuantitativas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Variables categóricas

### Gráficas de conteo y de barras

```python
# Importar conjunto de datos

df = pd.read_csv("Survey.csv", index_col=0)
```

```python
# Gráfica para el conteo de observaciones

plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
sns.countplot(y="Internet usage", data=df) # Conteo para uso de internet
plt.xticks(rotation=45) # Cambiar la orientación de los ejes

plt.subplot(2,1,2)
sns.countplot(x="Internet usage", data=df) # Conteo para uso de internet
plt.tight_layout()
plt.show()
```

```python
# La función catplot permite graficar diagramas de barras para clase e incluir intervalos de confianza

sns.catplot(x="Gender", y="Mathematics", data=df, kind="bar", order=["male","female"]) # Se muestra el valor promedio
plt.show()
```

### Diagramas de caja

```python
# Importar dataset de estudiantes
df_2 = pd.read_csv("Students.csv", index_col=0)
```

```python
# Para la función catplot se define en el eje x los grupos a los cuales se les desea realizar el diagrama de cajas

sns.catplot(x="study_time", y="G3", data=df_2, kind="box", palette="viridis")
plt.show()
```

```python
# Igualmente es posible distinguir al interir de un grupo diferentes sub grupos y hacer diagramas diferenciados

sns.catplot(x="internet", y="G3", data=df_2, kind="box", hue="location", sym="") # El argumento sym no muestra los atípicos
plt.show()
```

```python
# En la visualización y elaboración de los diagramas de caja se suele tomar 1.5 rangos inter cuartiles para graficar los
# "bigotes". Ahora, es posible modificar este valor con el argumento whis, así:

sns.catplot(x="internet", y="G3", data=df_2, kind="box") # 1.5 rangos
sns.catplot(x="internet", y="G3", data=df_2, kind="box", whis=0.5) # 0.5 rangos
sns.catplot(x="internet", y="G3", data=df_2, kind="box", whis=[5,95]) # percenti 5 y 95
sns.catplot(x="internet", y="G3", data=df_2, kind="box", whis=[0,100]) # Min y max
plt.show()
```

### Diagramas de intervalo (point plot)

```python
sns.catplot(x="famrel", y="absences", data=df_2, kind="point", capsize=0.2)
plt.show()
```

```python
# Como en los casos anteriores, es posible realizar gráficos para subgrupos o si es el caso cambiar el estimador

sns.catplot(x="romantic", y="absences", data=df_2, kind="point", hue="school", capsize=0.1)
plt.title("Intervalos para subgrupos")

from numpy import median
sns.catplot(x="romantic", y="absences", data=df_2, kind="point", hue="school", estimator=median, capsize=0.1)
plt.title("Mediana para subgrupos")
plt.show()
```
