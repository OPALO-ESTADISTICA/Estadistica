# Gr√°ficos para dos variables cuantitativas

```python
# Importar paquetes

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Students.csv") # Importar conjunto de datos

# Uso de graficos relacionales

sns.relplot(x="absences", y="G3", data=df, col="study_time") # Graficos relacionales para dos variables
plt.show()
```

```python
# Es posible diferencialos por filas y columnsas de acuerdo a ciertos grupos o factores. Por ejemplo, para este caso
# se considera diferenciar si el desempe√±o de los estudiantes difiere cuando reciben apoyo por parte de su familia o extra

sns.relplot(x="G1", y="G2", data=df, col="famsup", col_order=["yes","no"], row="schoolsup", row_order=["yes","no"], height=4,
           aspect=1.5)
plt.show()
```

## Personalizaci√≥n de gr√°ficos de dispersi√≥n

```python
df_2 = pd.read_csv("Automobile per gallon.csv")   # Importar un segundo conjunto de datos

# Es posible personalizar el tama√±o de los puntos en fuci√≥n del valor cuantitativo de la varible, igual que diferenciarlo por color
# Los argumentos height/aspect permiten modificar las dimensiones de los gr√°ficos

sns.relplot(y="mpg", x="horsepower", data=df_2, kind="scatter", size="cylinders", hue="cylinders", height=5, aspect=1)
plt.show()
```

```python
# Igualemnte es posible diferenciar los puntos como diferente figuras seg√∫n una categor√≠a

sns.relplot(y="mpg", x="horsepower", data=df_2, kind="scatter", hue="origin",
             style="origin", height=5, aspect=1)
plt.show()
```

## Gr√°ficos de l√≠neas

```python
df_2.columns
```

```python
# Se considera ahora la posibilidad de trazar l√≠neas para la relaci√≥n de dos variables

sns.relplot(x="model_year", y="mpg", data=df_2, kind="line")
plt.show()

# El √°rea sombreada indica el intervalo de confianza para ese a√±o
```

```python
# En lugar del intervalo se puede especificar para que los limites est√©n dados por las desviaciones est√°ndar

sns.relplot(x="model_year", y="mpg", data=df_2, kind="line",ci="sd")
plt.show()
```

```python
# Al igual que en el diagrama de dispersi√≥n, existe la posibilidad de diferenciar las observaciones por categor√≠a, en este caso,
# se toma por el estilo de las l√≠neas

sns.relplot(x="model_year", y="horsepower", data=df_2, kind="line", style="origin", hue="origin",
            ci=None, dashes=False, markers=True)
plt.show()
```
