# Introducción a seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Uso de pandas con seaborn

```python
# Importar conjunto de datos

df = pd.read_csv("Countries.csv", decimal=",")

gdp = df['GDP ($ per capita)']          # gross domestic product (Producto interno bruto)
phones = df['Phones (per 1000)']        # Número de teléfonos por cada 1000 habitantes
percent_literate = df['Literacy (%)']   # Porcentaje de la población alfabetizada
region = df['Region']                   # Región
```

```python
# Diagrama de dispersión para conjunto de datos

sns.scatterplot(x=gdp, y=phones)
plt.yticks()
plt.xticks(rotation=45)
plt.show()
```

```python
# Gráficos de conteo: Permiten visualizar la cantidad de veces que se presenta una variable categórica

sns.countplot(y=region)
plt.xlabel("Conteo")
plt.ylabel("Región")
plt.show()
```

```python
# Se importa un segundo conjunto de datos para evidenciar la asignación de colores y otros gráficos

df_2 = pd.read_csv("Students.csv")

sns.countplot(x="age", data=df_2)  # Conteo de los estudiantes por edad
plt.show()
```

```python
# Es posible hacer una diferencia por subgrupos dentro de la misma gráfica usando el argumento hue
# Igualmente, hue_order permite definir el orden de los subgrupos

plt.subplot(2,1,1)
sns.scatterplot(x="absences", y="G3", data=df_2, hue="location", hue_order=["Rural","Urban"])

plt.subplot(2,1,2)

colores = {"Rural":"green", "Urban":"blue"}
sns.scatterplot(x="absences", y="G3", data=df_2, hue="location", 
                hue_order=["Rural","Urban"], palette=colores)

plt.tight_layout()
plt.show()
```
