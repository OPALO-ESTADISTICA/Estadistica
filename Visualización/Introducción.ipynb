## Múltiples gráficos<a class="anchor" id="Cap1"></a>

En esta primera parte se emplea el dataset "Degree" que muestra las graduaciones de mujeres desde el año 1971 hasta 2011

```python
df = pd.read_csv("../Datos/Degree.csv", index_col=0)
```

```python
computer_science = df["Computer Science"]  # Se crea una variable para el atributo computer_science
physical_science = df["Physical Sciences"] # Se crea una variable para el atributo physical_science
year = df.Year
```

```python
plt.plot(year, computer_science, color='blue', label="Computer Science")
plt.legend()
plt.plot(year, physical_science, color='red', label="Physical Science")
plt.legend()
plt.show()
```

```python
# El comando axes permite definir los ejes de mi grafico

plt.axes([0.05, 0.05, 0.425, 0.7]) # plt.axes([xlo, ylo, width, height])
plt.plot(year, computer_science, color='blue', label="Computer Science")
plt.legend()

plt.axes([0.575, 0.05, 0.425, 0.7])
plt.plot(year, physical_science, color='red', label="Physical Science")
plt.legend()
plt.show()
```

```python
# El problema de graficar especificando los ejes es que se requiere probar para asignar de manera adecuada
# Otra opción es usando la función de subplot propia de matplotlib

plt.subplot(1,2,1) # plt.subplot([filas, columnas, ubicación])
plt.plot(year, computer_science, color='blue', label="Computer Science")
plt.legend()

plt.subplot(1,2,2)
plt.plot(year, physical_science, color='red', label="Physical Science")
plt.legend()

plt.tight_layout(pad=1.01) # Mejorar el espacio entre gráficos
plt.show()
```

## Personalización de ejes<a class="anchor" id="Cap2"></a>

```python
# Las funciones plt.xlim y plt.ylim permiten delimitar los ejes en el gráfico

plt.plot(year, computer_science, color='blue', label="Computer Science")
plt.plot(year, physical_science, color='red', label="Physical Science")
plt.xlim(1980,1990)
plt.ylim(0,80)
plt.legend()
plt.show()
```

```python
# Igualmente pueden ser delimitados con la función plt.axis([])

plt.axis([1980, 1990, 0, 80]) # las dimensiones de los ejes están dado por una lista
plt.plot(year, computer_science, color='blue', label="Computer Science")
plt.plot(year, physical_science, color='red', label="Physical Science")
plt.legend()
plt.show()
```

## Anotaciones y etiquetas<a class="anchor" id="Cap3"></a>

Aunque previamente ya se han empleado etiquetas, se resalta una vez más su uso y la posibilidad de incluir anotaciones dentro de los mismos gráficos para facilitar su presentación

```python
# Incluir etiquetas a los gráficos

plt.plot(year, computer_science, color='blue', label="Computer Science")
plt.legend()
plt.plot(year, physical_science, color='red', label="Physical Science")
plt.legend(loc= "lower center")

# Incluir una anotación

y_max = year[computer_science.argmax()] # Posición de la anotación
c_max = computer_science.max()          # Posición de la anotación

plt.annotate("Maximo", xy=(y_max, c_max), xytext=(y_max+5, c_max+5), 
             arrowprops={"color":'black',"arrowstyle":"->"})
plt.show()
```

Por defecto, matplotlib realiza los gáficos de manera similar a matlab, pero es posible configurar y utilizar otros estilos

```python
plt.style.available # Ver los estilos disponibles para graficar
plt.style.use('ggplot')

plt.plot(year, computer_science, color='blue', label="Computer Science")
plt.plot(year, physical_science, color='red', label="Physical Science")
plt.axis([1980, 1990, 0, 80])
plt.legend()
plt.show()
```
