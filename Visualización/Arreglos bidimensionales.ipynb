## Arreglos bidimensionales<a class="anchor" id="Cap1" ></a>

```python
u = np.linspace(-2, 2, 21)
v = np.linspace(-1, 1, 21)

# La función meshgrid permite crear un conjunto de coordenadas en base a dos vectores

X, Y = np.meshgrid(u, v) 
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 
plt.pcolor(Z)
plt.show()
```

```python
# La fucnión pcolor() permite graficar una matriz de coordenadas y asignarles un valor (color)
# Es usada para visualizar arreglos en 2D

A = np.array([[1,0,-1], [2,0,1], [1,1,1]])
plt.imshow(A, cmap="Blues")
plt.colorbar()
plt.show()
```

```python
A= np.array([[1,0,-1], [2,0,1], [1,1,1]])
plt.pcolor(A, cmap="Blues")
plt.colorbar()
plt.show()
```

### Gráficos de contorno

```python
Z = X**2 + np.sin(Y)*1.7

plt.contour(X, Y, Z, cmap="magma")
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
plt.contour(X, Y, Z)

plt.subplot(2,2,2)
plt.contour(X, Y, Z, 20)

plt.subplot(2,2,3)
plt.contourf(X, Y, Z)

plt.subplot(2,2,4)
plt.contourf(X, Y, Z,20)

plt.tight_layout()
plt.show()
```

## Construcción de diagramas bivariables<a class="anchor" id="Cap2" ></a>

```python
df = pd.read_csv("Automobile per gallon.csv",index_col=0)

mpg = df["mpg"]
hp = df["hp"]
```

```python
# Construir un histograma de dos dimensiones, cada color representa la frecuecia de un par

plt.hist2d(hp, mpg, bins=(20,20))  # bins establece la cantidad de clases por eje
plt.xlabel("Caballos de fuerza (hp)")
plt.ylabel("Millas por galón (mpg)")
plt.colorbar()
plt.show()
```

```python
# Es posible crear histogramas en dos dimensiones donde la frecuencia se representa en un hexágono

plt.hexbin(hp, mpg, gridsize=(15,12), extent=(40,235,8,48))  # bins establece la cantidad de clases por eje
plt.xlabel("Caballos de fuerza (hp)")
plt.ylabel("Millas por galón (mpg)")
plt.colorbar()
plt.show()
```

## Presentación de imágenes<a class="anchor" id="Cap3" ></a>

```python
# Importar una imágen en python

img = plt.imread("Astronauta.jpg") # Importar la imágen
print(img.shape) # Imprimir la forma del arreglo matricial

plt.imshow(img)  # Mostrar la imágen
plt.axis("off")  # No mostrar los ejes
plt.show()
```

```python
# Convertir la imágen a una escala de grises

intensity = img.sum(axis=2)  # Se suma el eje Z para obtener las intesidades de la imágen
plt.imshow(intensity, cmap="gray")
plt.axis("off")
plt.show()
```

```python
# Especificar la extensión y el aspecto 
plt.subplot(2,2,1)
plt.title('extent=(-1,1,-1,1),\naspect=0.5') 
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=0.5)

# # Especificar la extensión y el aspecto - superior derecha
plt.subplot(2,2,2)
plt.title('extent=(-1,1,-1,1),\naspect=1')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=1)

# # Especificar la extensión y el aspecto - inferior izquierda
plt.subplot(2,2,3)
plt.title('extent=(-1,1,-1,1),\naspect=2')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=2)

# # Especificar la extensión y el aspecto - inferior derecha
plt.subplot(2,2,4)
plt.title('extent=(-2,2,-1,1),\naspect=2')
plt.xticks([-2,-1,0,1,2])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-2,2,-1,1), aspect=2)

# Mejorar el espaciado de la figura
plt.tight_layout()
plt.show()
```
