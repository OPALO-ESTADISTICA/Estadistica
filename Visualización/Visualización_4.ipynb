## Visualización de series de tiempo <a class="anchor" appear id=Cap1></a>

```python
# Importar conjunto de datos de stock

df = pd.read_csv("stock.csv", index_col=1)
df = df.drop(df.columns[0], axis=1)         # Eliminar una columns
df.index = pd.to_datetime(df.index)         # Convertir los índices en tipo fecha 
```

Se toma como conjunto de datos el inventario de cuatro artículos para visualizar su comportamiento a lo largo del tiempo

```python
# Se grafican los cuatro artículos en un mísmo diagrama

plt.plot(df["AAPL"], color="blue", label="AAPL")
plt.plot(df["IBM"], color="green", label="IB.")
plt.plot(df["CSCO"], color="red", label="CSCO")
plt.plot(df["MSFT"], color="magenta", label="MSFT")
plt.xticks(rotation=60) # Rotar las etiquetas de los ejes

plt.legend(loc='upper left')
plt.show()
```

```python
# Es posible segmentar por trazos de gráfica para tener detalles de rangos de fechas

plt.subplot(2,1,1)
plt.plot(df["AAPL"], color="blue")

view = df["AAPL"]["2007":"2008"]  # Definir el rango de un año

plt.subplot(2,1,2)
plt.plot(view, color="black")
plt.title("AAPL: 2007-2008")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# En lugar de graficar dos gráficos a parte, es posible realizar un fragmento dentro del mismo gráfico

plt.figure(figsize=(8,6))
plt.plot(df["AAPL"], color="blue")
plt.xticks(rotation=45)
plt.title('AAPL: 2001-2011')

view = df["AAPL"]['2012-05':'2012-12']

# Especificar el lugar del gráfico interno
plt.axes([0.25,0.5,0.35,0.35])

plt.plot(view, color="red")
plt.xticks(rotation=45)
plt.title('2012/05-2012/12')
plt.show()
```
